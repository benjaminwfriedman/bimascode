# BIM as Code - Building Demo

This notebook demonstrates creating a building with levels and visualizing it.

```python
from bimascode import Building, Level
from build123d import Box, Location, Align
from ocp_vscode import show, reset_show, set_defaults
```

```python
# Configure viewer
set_defaults(reset_camera=True, helper_scale=5)
reset_show()
```

```python
# Create a building
building = Building(name="Demo Building", address="123 Main St")

# Add levels
ground = Level(building, "Ground Floor", 0)
first = Level(building, "First Floor", 4000)
second = Level(building, "Second Floor", 8000)

print(f"Created {building.name} with {len(building.levels)} levels")
```

```python
# Visualize as floor plates
for level in building.levels:
    floor = Box(10000, 15000, 200, align=(Align.CENTER, Align.CENTER, Align.MIN))
    floor = floor.move(Location((0, 0, level.elevation_mm)))
    show(floor, names=[level.name])
    print(f"Added {level.name} at elevation {level.elevation.m}m")
```

```python
# Export to IFC
building.export_ifc("demo_building.ifc")
print("✓ Exported to demo_building.ifc")
```
